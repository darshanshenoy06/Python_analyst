```python
gb = pd.DataFrame({
    "Employee":["John","Mike","Sarah","Tom","David","Emma"],
    "Department":["IT","HR","IT","Finance","HR","IT"],
    "Gender":["M","M","F","M","M","F"],
    "Salary":[50000,60000,70000,55000,65000,80000]
})
```

```python
gb.groupby('Department')['Salary'].sum()
```

```python
gb.groupby("Department")["Employee"].count()
```

```python
gb.groupby("Department")["Salary"].agg(
    ["sum","mean","max","min","count"]
)
```

```python
gb.groupby(["Department","Gender"])["Salary"].mean()
```

```python
gb.groupby(["Department","Gender"])["Salary"].mean().reset_index()
```

```python
gb.groupby("Department").agg(
    Mean_Salary=("Salary","mean"),
    Max_Salary=("Salary","max"),
    Employee_Count=("Employee","count")
)
```
